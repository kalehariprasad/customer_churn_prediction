```python
# Function to identify outliers based on IQR method
def find_outliers_iqr(df, num_cols):
    outliers = {}
    for col in num_cols:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range

        # Define the outlier cutoff points
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outlier_data = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outliers[col] = outlier_data

    return outliers

# Find outliers in the numeric columns
outliers = find_outliers_iqr(df, num_cols)

# Display outliers for each column
for col, outlier_data in outliers.items():
    print(f"Outliers in {col}:")
    print(outlier_data[col])
    print('-' * 40)
```



*   based on the both box plot and IQR method we can conclude we dont have any outlier in data
